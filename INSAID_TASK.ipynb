```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(df):
    vif = pd.DataFrame()
    vif['variables'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return(vif)

calc_vif(df)
```

We can see that oldbalanceOrg and newbalanceOrig have too high VIF thus they are highly correlated. Similarly oldbalanceDest and newbalanceDest. Also nameDest is connected to nameOrig.
Thus combine these pairs of collinear attributes and drop the individual ones.

```python
df['Actual_amount_orig'] = df.apply(lambda x: x['oldbalanceOrg'] - x['newbalanceOrig'],axis=1)
```

```python
df['Actual_amount_dest'] = df.apply(lambda x: x['oldbalanceDest'] - x['newbalanceDest'],axis=1)
```

```python
df['TransactionPath'] = df.apply(lambda x: x['nameOrig'] + x['nameDest'],axis=1)
```

```python
new_df = df.drop(['oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest','step','nameOrig','nameDest'],axis=1)
```

```python
calc_vif(new_df)
```

Using the VIF values and correlation heatmap. We just need to check if there are any two attributes highly correlated to each other and then drop the one which is less correlated to the isFraud Attribute.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
new_df["NormalizedAmount"] = scaler.fit_transform(new_df["amount"].values.reshape(-1, 1))
new_df.drop(["amount"], inplace= True, axis= 1)
```

```python
Y = new_df["isFraud"]
X = new_df.drop(["isFraud"], axis= 1)
```

I did not normalize the complete dataset because it may lead to decrease in accuracy of model.



```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

The reason i choose Random Forest Classifier and Decision Tree Classifier because models like XGBoost, Bagging, ANN, and Logistic Regression may give good accuracy but they won't give good precision and recall values.

```python
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y, test_size= 0.3, random_state= 42)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)

Y_pred_dt = decision_tree.predict(X_test)
```

```python
print("Classification Report:\n", classification_report(Y_test, Y_pred_dt))
print("ROC-AUC Score:", roc_auc_score(Y_test, Y_pred_dt))
sns.heatmap(confusion_matrix(Y_test, Y_pred_dt), annot=True, fmt="d")
plt.show()
```

```python
random_forest = RandomForestClassifier(n_estimators= 100)
random_forest.fit(X_train, Y_train)

Y_pred_rf = random_forest.predict(X_test)
```

print("Classification Report:\n", classification_report(Y_test, Y_pred_rf))
print("ROC-AUC Score:", roc_auc_score(Y_test, Y_pred_rf))

plt.show()

```python
decision_tree_score = decision_tree.score(X_test, Y_test) * 100
random_forest_score = random_forest.score(X_test, Y_test) * 100
```

```python
print("Decision Tree Score: ", decision_tree_score)
print("Random Forest Score: ", random_forest_score)
```

--In our fraud detection, precision is crucial because we want to correctly identify fraud cases while avoiding false accusations of innocent transactions. Although both Random Forest and Decision Tree models have similar accuracy, Random Forest has higher precision, making it better for fraud detection. This higher precision helps ensure we catch actual fraudsters without falsely flagging legitimate transactions, which is why these models are preferred over others..
